```python
# IQR method
plt.figure(figsize=(10, 5))
sns.boxplot(x = "CoapplicantIncome", data = data)
plt.show()
```

```python
# IQR = Q3 - Q1
''' 
min = Q1 - 1.5 * IQR
max = Q3 + 1.5 * IQR

'''
q1 = data["CoapplicantIncome"].quantile(0.25)
q3 = data["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - 1.5 * IQR
max_range = q3 + 1.5 * IQR
```

```python
(min_range, max_range)
```

```python
new_data = data[data["CoapplicantIncome"] <= max_range]
```

```python
new_data.shape
```

```python
sns.boxplot(x = "CoapplicantIncome", data = new_data)
plt.show()
```

```python
# Z-Score Method

''' 
z = (x - mean(x)) / std(x)

the data says that all the data after [mean(x)-(3*std(x)), mean(x)+(3*std(x))] are considered to be the outliers
'''
```

```python
sns.boxplot(x = "CoapplicantIncome", data = data)
plt.show()
```

```python
sns.distplot(data["CoapplicantIncome"])
plt.show()
```

```python
# by using theoretical method
min_range = data["CoapplicantIncome"].mean() - (3 * data["CoapplicantIncome"].std())
max_range = data["CoapplicantIncome"].mean() + (3 * data["CoapplicantIncome"].std())
```

```python
min_range, max_range
```

```python
new_data1 = data[data["CoapplicantIncome"] <= max_range]
```

```python
sns.boxplot(x = "CoapplicantIncome", data = new_data1)
plt.show()
```

```python
# by using actual z-score
z_score = (data["CoapplicantIncome"] - data["CoapplicantIncome"].mean()) / data["CoapplicantIncome"].std()
```

```python
data["z_score"] = z_score
```
